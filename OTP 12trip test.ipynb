```python
# Создаем обучающую выборку.
# Так как у нас нет ответов для датасета test, то приведем датасет train к виду test и продолжим работать с новым датасетом
X = train.drop(['goal1','goal25','goal23', 'goal1', 'goal24', 'goal22', 'goal21','userid', 'orderid'], axis=1)
X.head()
```

```python
# Разделяем выборку
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
def predict_proba(m, x_train, x_test, y_train, y_test):
    print(m)
    model = m
    model.fit(x_train, y_train)
    predictions = model.predict(x_test)
    probabilities = model.predict_proba(x_test)
    print('Precision: {:.2f}'.format(precision_score(y_test, predictions)))
    print('Recall: {:.2f}'.format(recall_score(y_test, predictions)))
    print('roc_auc: {:.2f}'.format(roc_auc_score(y_test, probabilities[:,1])))
    print('F1: {:.2f}'.format(f1_score(y_test, predictions)))
    fpr, tpr, threshold = roc_curve(y_test, predictions)
    i = np.arange(len(tpr)) 
    roc = pd.DataFrame({'tf' : pd.Series(tpr-(1-fpr), index=i), 'threshold' : pd.Series(threshold, index=i)})
    roc_t = roc.iloc[(roc.tf-0).abs().argsort()[:1]]
    print('Threshold:',list(roc_t['threshold']))
    
```

```python
predict_proba(LogisticRegression(random_state = 1), X_train, X_test, y_train, y_test)
```

```python
predict_proba(RandomForestClassifier(random_state = 1, max_depth = 7), X_train, X_test, y_train, y_test)
```

```python
predict_proba(GradientBoostingClassifier(random_state = 1), X_train, X_test, y_train, y_test)
```
